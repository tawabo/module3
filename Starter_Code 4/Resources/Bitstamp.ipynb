```python
import pandas as pd
from pathlib import Path
```

```python
csvpath = Path("../Resources/bitstamp.csv")
```

```python
bitstamp_df = pd.read_csv(csvpath, index_col="Timestamp", parse_dates=True, infer_datetime_format=True)
```

```python
bitstamp_df = bitstamp_df.fillna(0)
```

```python
bitstamp_df["Close"] = bitstamp_df["Close"].str.replace("$","")
```
